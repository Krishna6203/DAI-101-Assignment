```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import os

# Verify file existence and handle missing file
file_path = "/mnt/data/prices.csv"
if not os.path.exists(file_path):
    print(f"File not found: {file_path}. Please upload the file or check the path.")
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]  # Get the uploaded file name

# Load the dataset
df = pd.read_csv(file_path)

# Handle missing values by filling only numeric columns with column mean
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Display basic information about the dataset
print("Dataset Info:")
df.info()
print("\nFirst 5 rows:")
print(df.head())

# Univariate Analysis
print("\nDescriptive Statistics:")
print(df.describe())

# Proportional bar charts
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nProportion of {col}:")
    print(df[col].value_counts() / len(df))

# Histogram for numerical columns
sns.histplot(df, kde=True)
plt.title("Histograms of Numerical Features")
plt.show()

# Line chart for trends over time (if time column exists)
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date').plot(figsize=(12,6))
    plt.title("Trend Over Time")
    plt.show()

# Box plots to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title("Boxplots of Numerical Features")
plt.show()

# Multivariate Analysis
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Means and variances per group
if 'mode' in df.columns:
    print("\nMeans and Variances Per Group:")
    print(df.groupby('mode')[numeric_columns].agg(['mean', 'var']))

# Correlation heatmap
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=[np.number])  # Ensure only numeric columns are used
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Hinton diagram
plt.figure(figsize=(10, 10))
for i, col in enumerate(numeric_columns):
    for j, row in enumerate(numeric_columns):
        plt.gca().add_patch(plt.Rectangle((i, j), 1, 1, color='black', alpha=abs(numeric_df.corr().iloc[i, j])))
plt.xticks(range(len(numeric_columns)), numeric_columns, rotation=90)
plt.yticks(range(len(numeric_columns)), numeric_columns)
plt.title("Hinton Diagram")
plt.show()

# Standardizing variables
scaler = StandardScaler()
df_scaled = scaler.fit_transform(numeric_df)

# PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_scaled)
sns.scatterplot(x=pca_components[:,0], y=pca_components[:,1])
plt.title("Scatterplot of Principal Components")
plt.show()

# LDA (if target variable exists)
if 'mode' in df.columns:
    lda = LDA(n_components=1)
    lda_components = lda.fit_transform(df_scaled, df['mode'])
    sns.histplot(lda_components, kde=True)
    plt.title("Separation Achieved by LDA")
    plt.show()

# Scatter Matrix Plot
sns.pairplot(numeric_df)
plt.show()

# Q-Q plots
for col in numeric_columns:
    sm.qqplot(df[col], line='s')
    plt.title(f"Q-Q Plot of {col}")
    plt.show()

# T-tests & ANOVA (if categorical variables exist)
if 'mode' in df.columns:
    groups = [df[df['mode'] == val]['streams'] for val in df['mode'].unique()]
    f_stat, p_val = stats.f_oneway(*groups)
    print(f"ANOVA Results: F-stat = {f_stat}, p-value = {p_val}")

# Bivariate analysis (scatterplot for relationships)
if 'energy_%' in df.columns and 'danceability_%' in df.columns:
    sns.scatterplot(x=df['energy_%'], y=df['danceability_%'])
    plt.title("Scatterplot of Energy vs Danceability")
    plt.show()

print("\nKey Insights:")
print("1. Identify trends in numerical data distributions.")
print("2. Detect outliers and potential anomalies in data.")
print("3. Analyze correlations between numerical features.")
print("4. Principal Component Analysis shows variance in data.")
print("5. Linear Discriminant Analysis shows group separability.")
```
